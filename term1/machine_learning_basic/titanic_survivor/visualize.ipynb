```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("data/train.csv")
fig = plt.figure(figsize=(18,6))

#画出死亡和幸存比例
plt.subplot2grid((3,3),(0,0))
df.Survived.value_counts(normalize=True).plot(kind="bar",alpha=0.6)
plt.title("Survived")

#年老的比年轻的死的多
plt.subplot2grid((3,3),(0,1))
plt.scatter(df.Survived, df.Age, alpha=0.1)
plt.title("Age wrt Survived") 

#罗列乘客等级
plt.subplot2grid((3,3),(0,2))
df.Pclass.value_counts(normalize=True).plot(kind="bar",alpha=0.6)
plt.title("Class")

#观察乘客等级与年龄的关系
plt.subplot2grid((3,3),(1,0),colspan=2)
for x in [1,2,3]:
    df.Age[df.Pclass == x].plot(kind="kde")
plt.title("Class wrt Age")
plt.legend(["1st", "2nd", "3rd"])
plt.show()

#乘客从哪里登船的
plt.subplot2grid((3,3),(2,0),colspan=2)
df.Embarked.value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plt.title("Embarked")
#print(df.shape)
#print(df.count())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

female_color = "#FA0000"

df = pd.read_csv("data/train.csv")
fig = plt.figure(figsize=(18,6))

#画出死亡和幸存比例
plt.subplot2grid((3,4),(0,0))
df.Survived.value_counts(normalize=True).plot(kind="bar",alpha=0.6)
plt.title("Survived")

#画出男性死亡和幸存比例
plt.subplot2grid((3,4),(0,1))
df.Survived[df.Sex == "male"].value_counts(normalize=True).plot(kind="bar",alpha=0.6)
plt.title("Men Survived")

#画出女性死亡和幸存比例
plt.subplot2grid((3,4),(0,2))
df.Survived[df.Sex == "female"].value_counts(normalize=True).plot(kind="bar",alpha=0.6, color=female_color)
plt.title("Women Survived")

#对比存活者的性别
plt.subplot2grid((3,4),(0,3))
df.Sex[df.Survived == 1].value_counts(normalize=True).plot(kind="bar",alpha=0.6)
plt.title("Sex of Survived")

#对比存活者的社会等级
plt.subplot2grid((3,4),(1,0), colspan=4)
for x in [1,2,3]:
    df.Survived[df.Pclass == x].plot(kind="kde")
plt.title("Class wrt Survived")
plt.legend(("1st","2nd","3rd"))


#对比有钱男人的存活比例
plt.subplot2grid((3,4),(2,0))
df.Survived[(df.Sex == 'male') & (df.Pclass == 1)].value_counts(normalize=True).plot(kind="bar",alpha=0.6)
plt.title("Rich Men Survived")

plt.subplot2grid((3,4),(2,1))
df.Survived[(df.Sex == 'male') & (df.Pclass == 3)].value_counts(normalize=True).plot(kind="bar",alpha=0.6)
plt.title("Poor Men Survived")

#对比有钱女人的存活比例
plt.subplot2grid((3,4),(2,2))
df.Survived[(df.Sex == 'female') & (df.Pclass == 1)].value_counts(normalize=True).plot(kind="bar",alpha=0.6,color=female_color)
plt.title("Rich Women Survived")

plt.subplot2grid((3,4),(2,3))
df.Survived[(df.Sex == 'female') & (df.Pclass == 3)].value_counts(normalize=True).plot(kind="bar",alpha=0.6,color=female_color)
plt.title("Poor Women Survived")

plt.show()
```
